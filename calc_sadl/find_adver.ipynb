```python
import pickle
file_convcifar10 = "dsa_cifar10_convcifar10_pgd0.03_all_layers_path.p"
file_vgg = "dsa_cifar10_vgg_pgd0.03_all_layers_path.p"
file_alexnet = "dsa_SVHN_alexnet_pgd0.03_all_layers_path.p"

file_convcifar10_last_layer = "dsa_cifar10_convcifar10_pgd0.03_last_layer.p"
file_vgg_last_layer = "dsa_cifar10_vgg_pgd0.03_last_layer.p"
file_alexnet_last_layer = "dsa_SVHN_alexnet_pgd0.03_last_layer.p"
```

```python
with open(file_convcifar10, 'rb') as handle:
    dangers_convcifar10_all_layers_path = pickle.load(handle)
    
with open(file_vgg, 'rb') as handle:
    dangers_vgg_all_layers_path = pickle.load(handle)    

with open(file_alexnet, 'rb') as handle:
    dangers_alexnet_all_layers_path = pickle.load(handle)    
```

```python
with open(file_convcifar10_last_layer, 'rb') as handle:
    dangers_convcifar10_last_layer = pickle.load(handle)
    
with open(file_vgg_last_layer, 'rb') as handle:
    dangers_vgg_last_layer = pickle.load(handle)    

with open(file_alexnet_last_layer, 'rb') as handle:
    dangers_alexnet_last_layer = pickle.load(handle)    
```

```python
neurons_last_layer = [dangers_convcifar10_last_layer, dangers_vgg_last_layer, dangers_alexnet_last_layer]
neurons_all_layers = [dangers_convcifar10_all_layers_path, dangers_vgg_all_layers_path, dangers_alexnet_all_layers_path]
for index in range(3):
    def check(n, all_neurons):
        success = False
        for k in all_neurons.keys():
            if n not in all_neurons[k]:
                success = True
                return success, k
        return success, k

    ss = []
    ks = []
    for n in neurons_last_layer[index]:
        success, k = check(n, neurons_all_layers[index])
        if success:
            ss.append(n)
            ks.append(k)
# print(ss)

    print(f"help {len(ss)} from all {len(neurons_last_layer[index])}")
    print(ks)
```
